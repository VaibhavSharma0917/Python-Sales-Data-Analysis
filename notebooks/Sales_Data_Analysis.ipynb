# Exploratory Data Analysis

```python
category_sales = df.groupby("Category")["Sales"].sum()

category_sales
```

```python
category_profit = df.groupby("Category")["Profit"].sum()

category_profit
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

region_sales
```

```python
region_profit = df.groupby("Region")["Profit"].sum()

region_profit
```

```python
subcategory_sales = (
    df.groupby("Sub-Category")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

subcategory_sales
```

```python
subcategory_profit = (
    df.groupby("Sub-Category")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

subcategory_profit
```

```python
top_customers = (
    df.groupby("Customer ID")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers
```

```python
top_products = (
    df.groupby(["Product Name", "Category"])["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)

top_products
```

```python
top_profit_products = (
    df.groupby(["Product Name", "Category"], as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=False)
      .head(10)
      .reset_index(drop=True)
)

top_profit_products
```

```python
top_loss_products = (
    df.groupby(["Product Name", "Category"], as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=True)
      .head(10)
      .reset_index(drop=True)
)

top_loss_products
```

```python
top_customers_profit = (
    df.groupby("Customer ID", as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=False)
      .head(10)
      .reset_index(drop=True)
)

top_customers_profit
```

```python
monthly_sales = (
    df.groupby(df["Order Date"].dt.month_name(), as_index=False)["Sales"]
      .sum()
)

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_sales["Order Date"] = pd.Categorical(
    monthly_sales["Order Date"],
    categories=month_order,
    ordered=True
)

monthly_sales = monthly_sales.sort_values("Order Date").reset_index(drop=True)

monthly_sales
```

```python
monthly_profit = (
    df.groupby(df["Order Date"].dt.month_name(), as_index=False)["Profit"]
      .sum()
)

monthly_profit["Order Date"] = pd.Categorical(
    monthly_profit["Order Date"],
    categories=month_order,
    ordered=True
)

monthly_profit = monthly_profit.sort_values("Order Date")

monthly_profit
```

```python
discount_profit = (
    df.groupby("Discount", as_index=False)["Profit"]
      .mean()
      .sort_values(by="Discount")
)

discount_profit
```

```python
shipping_sales = (
    df.groupby("Ship Mode", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False).reset_index(drop=True)
)

shipping_sales
```

```python
segment_sales = (
    df.groupby("Segment", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
)

segment_sales
```

```python
segment_profit = (
    df.groupby("Segment", as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=False)
)

segment_profit
```

```python
top_cities = (
    df.groupby("City", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
      .head(10).reset_index(drop=True)
)

top_cities
```

```python
top_states = (
    df.groupby("State", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
      .head(10).reset_index(drop=True)
)

top_states
```

```python
correlation = df[["Sales", "Profit", "Discount"]].corr()

correlation
```

# Data Visualization

```python
import matplotlib.pyplot as plt

category_sales = (
    df.groupby("Category", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
)

plt.figure(figsize=(8,5))
plt.bar(category_sales["Category"], category_sales["Sales"])

plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

```python
category_profit = (
    df.groupby("Category", as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=False)
)

plt.figure(figsize=(8,5))

plt.bar(category_profit["Category"], category_profit["Profit"])

plt.title("Total Profit by Category")
plt.xlabel("Category")
plt.ylabel("Total Profit")

plt.tight_layout()
plt.show()
```

```python
region_sales = (
    df.groupby("Region", as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
)

plt.figure(figsize=(8,5))

plt.bar(region_sales["Region"], region_sales["Sales"])

plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

```python
region_profit = (
    df.groupby("Region", as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=False)
)

plt.figure(figsize=(8,5))

plt.bar(region_profit["Region"], region_profit["Profit"])

plt.title("Total Profit by Region")
plt.xlabel("Region")
plt.ylabel("Total Profit")

plt.tight_layout()
plt.show()
```

```python
top_products = (
    df.groupby(["Product Name", "Category"], as_index=False)["Sales"]
      .sum()
      .sort_values(by="Sales", ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))

plt.barh(top_products["Product Name"], top_products["Sales"])

plt.title("Top 10 Products by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product Name")

plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
top_profit_products = (
    df.groupby(["Product Name", "Category"], as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=False)
      .head(10)
)

plt.figure(figsize=(12,6))

plt.barh(top_profit_products["Product Name"], top_profit_products["Profit"])

plt.title("Top 10 Products by Profit")
plt.xlabel("Total Profit")
plt.ylabel("Product Name")

plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
loss_products = (
    df.groupby(["Product Name", "Category"], as_index=False)["Profit"]
      .sum()
      .sort_values(by="Profit", ascending=True)
      .head(10)
)

plt.figure(figsize=(12,6))

plt.barh(loss_products["Product Name"], loss_products["Profit"])

plt.title("Top 10 Loss-Making Products")
plt.xlabel("Total Profit (Loss)")
plt.ylabel("Product Name")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    monthly_sales["Order Date"],
    monthly_sales["Sales"],
    marker="o"
)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    monthly_profit["Order Date"],
    monthly_profit["Profit"],
    marker="o"
)

plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Total Profit")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    discount_profit["Discount"],
    discount_profit["Profit"]
)

plt.title("Discount vs Average Profit")
plt.xlabel("Discount")
plt.ylabel("Average Profit")

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(6,5))

sns.heatmap(
    correlation,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix")

plt.tight_layout()
plt.show()
```

```python
category_summary = (
    df.groupby("Category", as_index=False)[["Sales", "Profit"]]
      .sum()
)

plt.figure(figsize=(9,5))

x = range(len(category_summary))

plt.bar(x, category_summary["Sales"], width=0.4, label="Sales")
plt.bar([i + 0.4 for i in x], category_summary["Profit"], width=0.4, label="Profit")

plt.xticks([i + 0.2 for i in x], category_summary["Category"])

plt.title("Sales vs Profit by Category")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
region_summary = (
    df.groupby("Region", as_index=False)[["Sales", "Profit"]]
      .sum()
)

plt.figure(figsize=(9,5))

x = range(len(region_summary))

plt.bar(x, region_summary["Sales"], width=0.4, label="Sales")
plt.bar([i + 0.4 for i in x], region_summary["Profit"], width=0.4, label="Profit")

plt.xticks([i + 0.2 for i in x], region_summary["Region"])

plt.title("Sales vs Profit by Region")
plt.xlabel("Region")
plt.ylabel("Amount")
plt.legend()

plt.tight_layout()
plt.show()
```

#  Executive Business Report




## Executive Summary

This project analyzed a retail sales dataset using Python to evaluate sales performance, profitability, customer behavior, regional trends, and the impact of discounting. The analysis involved data cleaning, exploratory data analysis (EDA), and data visualization to identify key business opportunities and operational challenges. The findings provide actionable recommendations that can support strategic decision-making and improve overall business performance.

---

## Key Business Insights

1. **Technology** emerged as the best-performing category, generating the highest revenue as well as the highest overall profit, making it the primary contributor to business growth.

2. Although **Furniture** contributes significantly to total sales, its profitability remains considerably lower than other categories, indicating potential pricing or cost-management issues.

3. **Phones** represent the highest-selling product sub-category, reflecting strong and consistent customer demand.

4. **Copiers** generate the highest profit among all sub-categories, suggesting excellent profit margins and strong commercial value.

5. **Tables** are the largest loss-making products despite healthy sales volumes, highlighting an opportunity to review pricing, discounts, and procurement costs.

6. The **West** and **East** regions consistently outperform other regions in both sales and profitability, while the **Central** region generates relatively high sales but comparatively lower profit.

7. The analysis demonstrates that **higher sales do not necessarily translate into higher profitability**, emphasizing the importance of evaluating both revenue and profit simultaneously.

8. A **negative relationship between Discount and Profit** indicates that excessive discounting reduces profitability without proportionally improving business performance.

9. A relatively **small group of products and customers contributes a significant share of total revenue and profit**, suggesting the importance of customer retention and product prioritization strategies.

10. Overall, the analysis demonstrates that **data-driven decision making** can identify operational inefficiencies, optimize pricing strategies, and improve long-term business performance.

---

## Strategic Recommendations

- Continue investing in high-performing **Technology** products to maximize revenue and profitability.
- Review pricing, manufacturing costs, and discount policies for the **Furniture** category.
- Investigate the persistent losses associated with **Tables** and implement corrective pricing or sourcing strategies.
- Optimize discount policies by limiting high discounts on low-margin products.
- Increase marketing efforts and inventory allocation in high-performing regions such as the **West** and **East**.
- Develop targeted improvement strategies for underperforming regions to improve profitability.
- Strengthen customer retention programs for high-value customers contributing significant revenue.
- Monitor profit-based KPIs alongside sales metrics to support more informed business decisions.
